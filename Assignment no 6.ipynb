## Class

#### A collection of data (fields/ variables) and methods that operate on that data. OR It is a template or blue print about the capability of what an object can do.
#### Data/methods define the contents/capabilities of the instances (objects) of the class.
#### A class can be viewed as a factory for objects.
#### A class defines a recipe for its objects.

## Object

#### A thing that has identity, state, and behavior. It is an instantiation of a class. When class is defined, only the description for the object is defined. Therefore, no memory or storage is located.

## Attribute

#### Attributes are data stored inside a class or instance and represent the state or quality of the class or instance. In short, attributes store information about the instance.

## Behaviour

#### A class's behavior determines how an instance of that class operates OR objects interact with each other , and it is defined by the operations an object can perform.
